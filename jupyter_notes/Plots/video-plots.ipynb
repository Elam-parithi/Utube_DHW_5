```python
import pandas as pd
from dataget import *
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
query = """
SELECT v.video_id, v.video_name, v.view_count, AVG(c.sentiment) AS average_sentiment FROM videos v
LEFT JOIN comments c ON v.video_id = c.video_id
GROUP BY v.video_id ORDER BY v.view_count DESC LIMIT 5;
"""
```

```python
data_df = get_query(query)
```

```python
data_df
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(x='video_id', y='view_count', data=data_df, ax=axes[0])
axes[0].set_title('View Count per Video')
axes[0].set_ylabel('View Count')

sns.barplot(x='video_id', y='average_sentiment', data=data_df, ax=axes[1])
axes[1].set_title('Average Sentiment per Video')
axes[1].set_ylabel('Average Sentiment')

plt.tight_layout()
plt.show()
```
