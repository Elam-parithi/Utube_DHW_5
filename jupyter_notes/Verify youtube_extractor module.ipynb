# Get the channel ID

# Get Channel Info

```python
channel_info = yt.get_channel_info(channel_ID)
channel_info
```

```python
uploads_id = channel_info['contentDetails']['relatedPlaylists']['uploads']
uploads_id
```

# Get all the playlist ID from that Channel

```python
playlist_response = yt.get_playlists_from_channel(channel_ID)
playlist_response
```

```python
len(playlist_response)
```

```python
playlist_list = [uploads_id]
```

```python
for item in playlist_response:
    playlist_list.append(item['id'])
    print(item['snippet']['title'])
```
